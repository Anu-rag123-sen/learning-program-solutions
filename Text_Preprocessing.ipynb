```python
#REMOVE URL
```

```python
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'', text)
```

```python
#PUNCTUATION HANDLING
```

```python
import string,time
string.punctuation
```

```python
exclude=string.punctuation
exclude
```

```python
def remove_punc(text):
  for char in exclude:
    text=text.replace(char, '')
  return text
```

```python
text='String . with . punctuation?'
```

```python
start=time.time()
print(remove_punc(text))
time1=time.time()- start
print(time1*50000)
```

```python
#you can also use text.translate function in order to remove punctuations
```

```python
def remove_punc1(text):
  return text.translate(str.maketrans('', '', exclude))
```

```python
start=time.time()
print(remove_punc1(text))
time2=time.time()- start
print(time2*50000)
```

```python
#CHAT CONVERSION HANDLDING
```

```python
chat_words = {
    'AFK': 'Away From Keyboard',
    'ASAP': 'As Soon As Possible',
    'HON': 'Human Onboarding'
}
```

```python
def chat_conversion(text):
  new_text=[]
  for w in text.split():
    if w.upper() in chat_words:
      new_text.append(chat_words[w.upper()])
    else:
      new_text.append(w)
  return " ".join (new_text)
```

```python
chat_conversion('Do this work ASAP')
```

```python
#function to remove stopwords from a text
```

```python
def remove_stopwords(text):
  new_text=[]
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x=new_text[:]
  new_text.clear()
  return " ".join(x)
```

```python
remove_stopwords('probably my all-time favorite movie, a story of selflessness, sacrifice and dedication to a noble cause, but it\'s not preachin anythin')
```

```python
#REMOVE EMOJI HANDLES
```

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & pictographs
        "\U0001F680-\U0001F6FF"  # Transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # Flags
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251"  # Enclosed characters
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r' ', text)
```

```python
remove_emojis("Loved the movie. It was ðŸ˜˜ðŸ˜˜")
```

```python
#STEMMER
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = 'walk walks walking walked'
stem_words(sample)
```

```python
text = 'I am meeting him tomorrow at the meeting'
stem_words(text)
```

```python
#LEMMATIZATION(IMPORTANT)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
wordnet_lemmatizer = WordNetLemmatizer()

text= 'I am meeting him tomorrow at the meeting'
punctuation= '?:!.,;'
sentence_words = nltk.word_tokenize(text)
for word in sentence_words:
  if word in punctuation:
    sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
  print("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word)))
```
