```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)
# Line Plot
plt.plot(x, y,label="Sin(x)")
plt.legend() # makes plot readable by showing labels
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
# Scatter Plot
plt.scatter(x, y, color="red", marker="o")
plt.title("Scatter Plot Example")
plt.show()
# Bar Plot
categories = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]
plt.bar(categories, values, color="green")
plt.title("Bar Plot Example")
plt.show()
# Histogram
data = np.random.randn(1000)  # 1000 random numbers
plt.hist(data, bins=30, color="purple", edgecolor="black") # The bins parameter decides how many intervals you want to divide your data range into.
plt.title("Histogram Example")
plt.show()
```

```python
# INTERMEDIATE LEVEL PLOTS
```

```python
# Multiple lines
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)", linestyle="--")
plt.legend()
plt.title("Multiple Line Plot")
plt.show()
# Stack Plot
days = [1,2,3,4,5]
eating = [2,3,4,3,2]
sleeping = [7,8,6,11,7]
working = [7,8,7,2,2]
plt.stackplot(days, eating, sleeping, working, labels=['Eat','Sleep','Work'])
plt.legend(loc='upper left')
plt.title("Stack Plot Example")
plt.show()
# Pie Chart
sizes = [30, 20, 25, 25]
labels = ["Python", "C++", "Java", "Ruby"]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()
# autopct='%1.1f%%'
#This controls the format of the percentage labels on the pie chart.
#Each slice shows its percentage value according to this format string.
#Breakdown of '%1.1f%%':
#%  placeholder for a value.
#1.1f  floating-point number with 1 digit before decimal and 1 after decimal.
#%%  literal percent sign % in the output.

#startangle=90
#This rotates the pie chart, setting the starting angle (in degrees) of the first slice.
#By default, the first slice starts at 0 (pointing to the right, along the x-axis).
#With startangle=90, the first slice starts at the top (12 oclock position).
```

```python
# ADVANCED LEVEL PLOTS
```

```python
# Subplots (Multiple graphs in One Figure)
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
axs[0,0].plot(x, np.sin(x)); axs[0,0].set_title("sin(x)")
axs[0,1].plot(x, np.cos(x), 'r'); axs[0,1].set_title("cos(x)")
axs[1,0].scatter(x, np.sin(x)); axs[1,0].set_title("Scatter")
axs[1,1].bar(categories, values); axs[1,1].set_title("Bar")
plt.tight_layout()
plt.show()
# Error Bar
y_err = 0.1 + 0.1 * np.sqrt(x)
plt.errorbar(x, y, yerr=y_err, fmt='o', ecolor='red', capsize=3)
plt.title("Error Bar Plot")
plt.show()
# 3D Plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
z = np.linspace(0, 1, 100)
x = z * np.sin(25*z)
y = z * np.cos(25*z)
ax.plot3D(x, y, z, 'green')
plt.title("3D Line Plot")
plt.show()
```
